pertama import numpy untuk komputasi matrix, matplot lib untuk gambar grafik, pandas untuk mengimport csv dan menganalisis data.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

selanjutnya mengimport dataset.

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

kemudian dataset dibagi menjadi x untuk menampung atribut dan y untuk label.

```python
x = dataset.iloc[:, [2,3]].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

kemudian dibagi menjadi data training dan data testing menggunakan train test split dari library sklearn.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
print(x_train)
```

kemudian karena interval antara Age dan Estimated Gaji jauh, maka perlu di scaling menggunakan StandardScaler.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

kemudian dilakukan modeling menggunakan KNN.

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski', p=2)
classifier.fit(x_train, y_train)
```

kemudian menentukan prediksi.

```python
y_pred = classifier.predict(x_test)
```

kemudian hitung evaluasinya menggunakan confusion matrix.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

kemudian menggambarkan grafik dari knn.

```python
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start=x_set[:,0].min()-1, stop=x_set[:,0].max()+1, step = 0.01),
                     np.arange(start=x_set[:,1].min()-1, stop=x_set[:,1].max()+1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate (np.unique(y_set)):
  plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
              c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Klasifikasi Data dengan K-NN (Data Training)')
plt.xlabel('Umur')
plt.ylabel('Estimasi Gaji')
plt.legend()
plt.show()
```
